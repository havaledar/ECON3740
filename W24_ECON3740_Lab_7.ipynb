# Reading data

## Downloading

```python
url = 'https://drive.google.com/file/d/1m0d9gZZBfHm7qGerKKLN7UzWaNUbQi0R/view?usp=sharing'
output_filename = 'lfs.dta'
gdown.download(url, output_filename, fuzzy=True ,quiet=False)
```

## Reading

```python
LFS = pd.read_stata('lfs.dta', convert_categoricals=True)
```

# Filter

```python
LFS = LFS[(LFS['HRLYEARN'] > 0) &
          (((LFS['COWMAIN'] == 'Private sector employees') |
           (LFS['COWMAIN'] == 'Public sector employees')
          ))]
```

# Wage ~ Education + Sex + Class of Worker

```python
m6 = smf.ols('HRLYEARN ~ EDUC + SEX + COWMAIN', LFS)

results_m6 = m6.fit()

Stargazer([results_m6])
```

## Renaming the covariants

```python
m6 = smf.ols('HRLYEARN ~ EDUC + SEX + COWMAIN', LFS)

results_m6 = m6.fit()

stargazer = Stargazer([results_m6])
stargazer.rename_covariates({'COWMAIN[T.Private sector employees]': 'Class of worker: Private sector employees',
                             'COWMAIN[T.Self-employed incorporated, no paid help]': 'Class of worker: Self-employed incorporated, no paid help',
                             'COWMAIN[T.Self-employed incorporated, with paid help]': 'Class of worker: Self-employed incorporated, with paid help',
                             'COWMAIN[T.Self-employed unincorporated, no paid help]': 'Class of worker: Self-employed unincorporated, no paid help',
                             'COWMAIN[T.Self-employed unincorporated, with paid help]': 'Class of worker: Self-employed unincorporated, with paid help'})
stargazer
```
