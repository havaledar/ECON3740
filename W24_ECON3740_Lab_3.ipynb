## Group by

Summary statistics by group

Collapsing the data by group is the easiest way of getting stats for continuous vars


````python
# Compute mean of 'AGE_6' grouped by 'SEX'

dMEAN = d.groupby('SEX').agg(AGE_6_mean=('AGE_6', 'mean'))
dMEAN
````

## Basic Regression Analysis

````python
# Drop missing values from the columns used in the regression
# data_clean = data.dropna(subset=['HRLYEARN', 'TENURE'])

# Create the model
model = smf.ols('HRLYEARN ~ TENURE', data)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
````

````python
# Create the model
model = smf.ols('np.log(HRLYEARN) ~ TENURE', data)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
````

````python
data['EDUC'].unique()
````

````python
# Create the model with EDUC as a categorical variable
model_categorical = smf.ols('HRLYEARN ~ C(EDUC)', data)

# Fit the model
results_categorical = model_categorical.fit()

# Display regression results
print(results_categorical.summary())
````
