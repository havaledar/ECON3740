## read_stata

Once the file is uploaded and the libraries you are going to use imported, you can read it into a DataFrame.

```python
data = pd.read_stata('file.dta')
```

Make sure the filename matches the uploaded file.

You can also reference the full Google Colab filesystem path if needed:

```
data = pd.read_stata('/content/file.dta')
```

For example, we can import a dta file of survey data downloaded from a website:

To read a dta file into a DataFrame:

```python
data = pd.read_stata('LFS-71M0001-E-2023-August_F1.dta')
```

Pandas will automatically convert the data into a DataFrame.

## Labels or Values

When you are reading the dta files, you can indicate if you need the values(convert_categoricals = False) or the lables(convert_categoricals = True) for the categorical variables as below. The read_stata by default read the lables not the values.

```python
data = pd.read_stata('LFS-71M0001-E-2023-August_F1.dta', convert_categoricals=False)
```

````python
data_labels = pd.read_stata("LFS-71M0001-E-2023-August_F1.dta")
data_labels
````
