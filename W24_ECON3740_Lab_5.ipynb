# Matplotlib

## Basics of matplotlib and its plotting functions
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It's highly customizable and works well with numpy and pandas data structures, making it a popular choice for data visualization in scientific computing.

## Setting up the environment
First, let's import the matplotlib library along with numpy for some basic data manipulation:

```python
import matplotlib.pyplot as plt
```

## Type of plots

### Line Plot
A basic plot to display data points linearly connected by straight line segments.

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.show()
```

### Scatter Plot
Used to plot points on a horizontal and vertical axis to show how much one variable is affected by another.

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)

plt.show()
```

### Bar Chart
A bar chart represents data with rectangular bars with lengths proportional to the values they represent.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 23, 6, 18]

plt.bar(categories, values)

plt.show()
```

###Histogram
A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.show()
```

### Pie Chart
A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportion.

```python
sizes = [25, 35, 20, 20]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

plt.pie(sizes, labels=labels)

plt.show()
```

### Box Plot
A box plot, or box-and-whisker plot, shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.

```python
data = np.random.rand(10, 4)

plt.boxplot(data)

plt.show()
```

### Heatmap
A heatmap contains values represented as colors and is used for visualizing the correlation matrix, among other uses.

```python
# Generating random data
data = np.random.rand(10, 10)

# Creating the heatmap
plt.imshow(data, cmap='hot', interpolation='nearest')

plt.show()
```

## CEOSAL plot
Now, we'll create a simple line plot.

```python
# Create x and y variables
x = data['ceoten']
y = data['salary']

# Create a line plot
plt.scatter(x, y)
# Display the plot
plt.show()
```

## Customizing plots
Matplotlib plots are highly customizable. You can change the labels, legends, colors, and markers to make your plots more informative and visually appealing.

### Figure size

```python
# Create a figure
plt.figure(figsize=(10, 8))

# Create x and y variables
x = data['ceoten']
y = data['salary']

# Create a line plot
plt.scatter(x, y)
# Display the plot
plt.show()
```

### Adding labels and title

```python
plt.scatter(x, y)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Color

```python
plt.scatter(x,
            y,
            color='green')

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Marker

```python
plt.scatter(x,
            y,
            color='green',
            marker='*')

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Size

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Transparency

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100,
            alpha=0.5)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Border

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100,
            alpha=0.5,
            edgecolors='black',
            linewidths=0.5)

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

plt.show()
```

### Adding a legend

```python
plt.scatter(x,
            y,
            color='green',
            marker='*',
            s=100,
            alpha=0.5,
            edgecolors='black',
            linewidths=0.5,
            label='Salary')

# Adding a title
plt.title('Salary in $1000s vs. Number of years in the Company as CEO')

# Adding the x and y labels
plt.xlabel('Years')
plt.ylabel('Salary')

# Adding the legend
plt.legend()

plt.show()
```

## Multiple plots

```python
z = data['mktval']

plt.scatter(x, y, label='Salary')

# Adding a second plot
plt.scatter(x, z, label='Market Value', color='red')

plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Salary and Market Value vs. Year')
plt.legend()

plt.show()
```

## Subplot

```python
# Create first subplot
plt.subplot(2, 1, 1)
plt.scatter(x, y, label='Salary')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Salary vs. Year')
plt.legend()

# Create second subplot
plt.subplot(2, 1, 2)
plt.scatter(x, z, label='Market Value', color='red')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Market Value vs. Year')
plt.legend()

# Adjust layout
plt.tight_layout()

plt.show()
```

## Regression line

### Statsmodel

```python
import statsmodels.formula.api as smf

m2 = smf.ols('salary ~ ceoten', data)

results = m2.fit()

intercept, slope = results.params

y_pred = slope * x + intercept
```

```python
# Create the scatter plot
plt.scatter(x, y, label='Data Points')

# Add the regression line
plt.plot(x, y_pred, color='red', label='Regression Line')

# Customize the plot
plt.xlabel('CEO Tenure')
plt.ylabel('Salary')
plt.title('CEO Tenure vs. Salary')
plt.legend()

# Display the plot
plt.show()
```
